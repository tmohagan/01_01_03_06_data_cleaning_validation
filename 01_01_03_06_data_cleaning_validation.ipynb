```python
df_oaprices.columns = ['pmid/pmcid', 'publisher', 'journal title', 'article title', 'cost']
```

```python
df_oaprices['journal title'] = df_oaprices['journal title'].str.lower()
df_oaprices['cost'] = df_oaprices['cost'].str.replace('£', '')
df_oaprices['cost'] = df_oaprices['cost'].str.replace('$', '')
df_oaprices['cost'] = df_oaprices['cost'].astype(float)
```
